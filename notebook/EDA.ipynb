#Attrition Count Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/hr_data.csv")

sns.countplot(x='Attrition', data=df)
plt.title("Attrition Count")
plt.show()
```

#Salary vs Attrition

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Salary vs Attrition")
plt.show()
```

#Job Role vs Attrition

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.title("Job Role vs Attrition")
plt.show()
```

#Overtime vs Attrition

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Overtime vs Attrition")
plt.show()
```

#Correlation Heatmap

```python
import numpy as np

plt.figure(figsize=(12,10))
sns.heatmap(df.select_dtypes(include=np.number).corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#Department vs Attrition (Advanced Analysis)

```python
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()

dept_attrition.plot(kind='bar', stacked=True)
plt.title("Attrition Rate by Department")
plt.ylabel("Percentage")
plt.show()
```

#Age Distribution

```python
sns.histplot(data=df, x='Age', hue='Attrition', kde=True)
plt.title("Age Distribution vs Attrition")
plt.show()
```

#Years at Company

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```

#Pairplot

```python
sns.pairplot(df[['Age','MonthlyIncome','YearsAtCompany','Attrition']], hue='Attrition')
plt.show()
```
